```python
from bs4 import BeautifulSoup
import requests

url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')

soup.find('table', class_='wikitable sortable')
table = soup.find_all('table')[0] #gets first instance of table, where main table is there

table_headers = table.find_all('th') #gets all table column headers for first table
table_headers_list = [header.text.strip() for header in table_headers] #gets plain text for each table column header and appends to the list
print(table_headers_list)
```

```python
import pandas as pd
df = pd.DataFrame(columns = table_headers_list)
```

```python
table_rows = table.find_all('tr') #find all rows.

for individual_row in table_rows[1:]: #This process makes 101 (100 after slice) lists. Each list is for its respective row and contains the respective data values.
    individual_row_data = individual_row.find_all('td') #finds all data for each individual row
    individual_row_data_list = [data.text.strip() for data in individual_row_data] #each and every data value in an individual row is put into a list
    #print(individual_row_data_list) - first row is the column headers row, where there is no 'td-class'/data to extract, so you get an empty list.
    #You have to get rid of the empty list for the below to work, so we add the slicer in the for loop
    length = len(df) #counts number of rows in df
    df.loc[length] = individual_row_data_list #populates each new row with relevant data
```

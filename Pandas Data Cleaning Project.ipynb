# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Ankur\OneDrive\Documents\data analysis\python stuffs\Customer Call List.xlsx")
df = df.set_index('CustomerID')
df
```

```python
df = df.drop_duplicates() #df = ... saves the query 
df
```

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("123./_")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex=True) #replaces everything (but a-z, A-Z and 0-9) with nothing
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x)) #converts all values to string
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10]) #applies formatting using concat and slicing
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--', '')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--', '')
df
```

```python
df[["Street Address", "State", "Zip-Code"]] = df["Address"].str.split(pat=',', n=2, expand=True)
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N')
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N')
df
```

```python
df = df.replace('N/a', '')
df = df.fillna('') #for Nan
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace = True)
#for each row (x) in the the df, if the row under Do_Not_Contact has Y, drop it
df
```

```python
for y in df.index:
    if df.loc[y, "Phone_Number"] == '':
        df.drop(y, inplace = True)
#for each row (y) in the the df, if the row under Phone_Number is empty, drop it
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('NNNNNNN', 'N')
df
```
